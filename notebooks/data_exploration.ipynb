# Data Exploration - Doctor Recommendation Dataset
This notebook explores the dataset used by the recommendation API.

```python
from pathlib import Path
import pickle as pk
import pandas as pd

BASE_DIR = Path.cwd().resolve().parent
MODEL_PATH = BASE_DIR / 'models' / 'doctor_ai_full_package.pkl'
DATA_PATH = BASE_DIR / 'data' / 'Dr.Seba_500_Organized_Final.xlsx'

with open(MODEL_PATH, 'rb') as f:
    package = pk.load(f)

le_spec = package['le_spec']
features = package['features']
df = pd.read_excel(DATA_PATH)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print('=' * 100)
print('DATASET EXPLORATION')
print('=' * 100)
print(f'\nDataset shape: {df.shape}')
print(f'\nColumn names: {list(df.columns)}')
```

```python
print('\n' + '=' * 100)
print('Available Districts:')
print('=' * 100)
print(df['district'].value_counts())

print('\n' + '=' * 100)
print('Available Specializations:')
print('=' * 100)
print(df['specialization_group'].value_counts())
```

```python
print('\n' + '=' * 100)
print('Sample Data (first 5 rows):')
print('=' * 100)
print(df[['doctor_name', 'district', 'thana', 'specialization_group', 'consultation_fees', 'hospital_name']].head())

print('\n' + '=' * 100)
print('Available Doctors in Barisal, Bakerganj:')
print('=' * 100)
barisal_data = df[(df['district'] == 'Barisal') & (df['thana'] == 'Bakerganj')]
print(f'Count: {len(barisal_data)}')
if len(barisal_data) > 0:
    print(barisal_data[['doctor_name', 'specialization_group', 'consultation_fees', 'hospital_name']].to_string(index=False))
```

```python
print('\n' + '=' * 100)
print('Doctors with Surgeon specialization (Encoded value):')
print('=' * 100)
surgeon_encoded = le_spec.transform(['Surgeon'])[0]
print(f'Surgeon encoded value: {surgeon_encoded}')
surgeon_doctors = df[df['specialization_group'] == surgeon_encoded]
print(f'Total Surgeons: {len(surgeon_doctors)}')
print(surgeon_doctors[['doctor_name', 'district', 'thana', 'consultation_fees']].head(10).to_string(index=False))
```
